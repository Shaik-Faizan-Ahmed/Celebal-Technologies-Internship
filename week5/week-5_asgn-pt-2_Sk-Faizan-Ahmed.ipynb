# Q1. What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?

## Answer

Traditional MapReduce is a disk-based processing framework where intermediate results are written to disk after every Map and Reduce stage. This increases execution time, especially for large datasets and iterative applications.

### Limitations of MapReduce

- High disk I/O because intermediate data is stored on disk.
- Slower execution due to repeated disk reads and writes.
- Poor performance for iterative algorithms such as machine learning.
- Higher latency, making it unsuitable for real-time analytics.
- More complex programming model requiring separate Map and Reduce functions.
- Limited support for interactive data processing.

### Why Spark is Preferred

Apache Spark overcomes these limitations by using in-memory computing, which significantly reduces disk I/O and improves processing speed. Spark also provides easy-to-use APIs, supports batch and streaming workloads, machine learning, graph processing, and interactive analytics within a single framework.

# Q2. Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.

## Answer

Apache Spark uses **in-memory computing**, which means it stores intermediate data in memory (RAM) instead of writing it to disk after every processing stage. This significantly reduces the time spent on disk read and write operations.

In iterative machine learning algorithms, the same dataset is processed multiple times to improve the model. In traditional disk-based systems such as MapReduce, the dataset is read from disk during every iteration, which increases execution time.

Spark overcomes this limitation by keeping the data in memory across multiple iterations. Since the data is readily available in RAM, Spark can repeatedly access it without performing expensive disk operations. This results in much faster execution and improved performance for machine learning, graph processing, and other iterative workloads.

### Advantages of In-Memory Computing

- Reduces disk I/O operations.
- Speeds up iterative computations.
- Improves the performance of machine learning algorithms.
- Supports faster interactive data analysis.
- Enables efficient processing of large datasets.

### Conclusion

By storing intermediate data in memory, Apache Spark minimizes disk access and significantly improves the execution speed of iterative algorithms compared to traditional disk-based systems such as MapReduce.

# Q4. Given a DataFrame `df_sales`, write a query to filter for rows where the region is **'West'** and then group by **product_category** to find the average **sale_amount**.

## Solution

The dataset is first filtered to include only records where the region is **West**. The filtered data is then grouped by **product_category**, and the average **sale_amount** is calculated for each category using the `avg()` aggregation function.

# Q5. What is the difference between `.na.drop()` and `.na.fill()`? Provide a code example of filling null values in a `status` column with the string `'Unknown'`.

## Answer

Spark provides different methods to handle missing values in a DataFrame.

- **`.na.drop()`** removes rows that contain null values.
- **`.na.fill()`** replaces null values with a specified value without removing any rows.

The choice depends on the dataset. If missing values are important, `.na.fill()` is preferred because it preserves all records. If incomplete records are not useful, `.na.drop()` can be used to remove them.

## Solution

The following code replaces all null values in the **status** column with the string **"Unknown"**.

````python
# Filling null values in status column
df_status = df.na.fill({
    "status": "Unknown"
})
display(df_status)
````

````python
# Verifying null values
from pyspark.sql.functions import col, count, when
df_status.select(
    count(when(col("status").isNull(), True)).alias("Status Nulls")
).show()
````

# Q7. How does the immutability of Spark DataFrames affect how you perform "data cleaning" steps like dropping columns or renaming them?

## Answer

Spark DataFrames are **immutable**, which means they cannot be modified after they are created. Whenever a data cleaning operation such as dropping a column, renaming a column, or filtering records is performed, Spark creates a **new DataFrame** instead of changing the existing one.

For example, if a column is renamed using `withColumnRenamed()` or removed using `drop()`, the original DataFrame remains unchanged. The result must be stored in a new DataFrame or assigned back to the same variable.

### Example

```python
df_new = df.withColumnRenamed("raw_timestamp", "event_time")
```

In this example, `df_new` contains the updated column name, while the original `df` DataFrame remains unchanged.

### Advantages of DataFrame Immutability

- Prevents accidental modification of the original dataset.
- Makes data processing more reliable and consistent.
- Allows Spark to optimize execution using lazy evaluation.
- Supports fault tolerance and efficient parallel processing.

### Conclusion

DataFrame immutability ensures that every transformation creates a new DataFrame, making Spark applications safer, more reliable, and easier to optimize.

# Q11. Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?

## Answer

A **shuffle** is the process in Apache Spark where data is redistributed across different partitions during certain operations such as `groupBy()`, `join()`, and `orderBy()`.

During a grouping operation, records with the same key may exist in different partitions. Spark moves these records to the same partition before performing the aggregation. This movement of data between partitions is called a **shuffle**.

A shuffle is considered a **wide transformation** because data from multiple input partitions is combined to produce the output. Unlike narrow transformations, wide transformations require communication between partitions, which increases execution time.

### Characteristics of Shuffle Operations

- Redistributes data across partitions.
- Occurs during operations such as `groupBy()`, `join()`, and `orderBy()`.
- Involves network communication between partitions.
- Requires additional disk and memory resources.
- Increases execution time compared to narrow transformations.

### Conclusion

Shuffle operations are necessary for grouping and aggregation tasks, but they are more expensive because Spark must move data between partitions before processing. Therefore, grouping operations are classified as wide transformations.

# Q12. Write a code snippet that identifies and removes rows where the `email` column contains null values OR the `username` is an empty string.

## Solution

The following code first identifies records where the **email** column contains null values or the **username** column contains empty strings. It then removes these inconsistent records to improve the quality of the dataset.

````python
# Identifying inconsistent records
invalid_records = df.filter(
    (col("email").isNull()) |
    (trim(col("username")) == "")
)
display(invalid_records)
````

````python
# Removing inconsistent records
clean_df = df.filter(
    col("email").isNotNull()
).filter(
    trim(col("username")) != ""
)
display(clean_df)
````

````python
# Verifying the cleaned dataset
print("Rows Before Cleaning:", df.count())
print("Rows After Cleaning:", clean_df.count())
````

# Q14. In the context of cleaning a dataset, what is the risk of using `inferSchema=true` when your source data contains messy or inconsistent date formats?

## Answer

The `inferSchema=true` option allows Spark to automatically detect the data type of each column while reading a dataset. Although this is convenient, it can lead to incorrect data type detection when the dataset contains inconsistent or messy date formats.

For example, if some records use the format **YYYY-MM-DD** while others use **DD/MM/YYYY**, Spark may infer the column as a **string** instead of a **date** or **timestamp**. This can lead to incorrect sorting, filtering, and date-based calculations.

### Risks of Using `inferSchema=true`

- Incorrect data type detection.
- Date columns may be read as strings.
- Errors during filtering or date calculations.
- Inconsistent results when processing data.
- Additional cleaning may be required before analysis.

### Best Practice

When working with important date or timestamp columns, it is better to define the schema explicitly or convert the column to the required data type after loading the dataset using functions such as `cast()` or `to_timestamp()`.

### Conclusion

While `inferSchema=true` simplifies data loading, inconsistent date formats can result in incorrect schema inference. Therefore, validating or explicitly defining the schema is recommended when working with messy datasets.

# Q15. Write a final processing pipeline that:

1. Filters out duplicates.
2. Fills null prices with `0`.
3. Groups by `store_id` to calculate total revenue.

## Solution

The following Spark pipeline combines multiple DataFrame transformations into a single workflow. It removes duplicate records, replaces null values in the **price** column with **0**, and groups the dataset by **store_id** to calculate the total revenue generated by each store.

````python
# Building the processing pipeline
pipeline_df = (
    df
    .dropDuplicates(["user_id", "transaction_date"])
    .fillna({
        "price": 0
    })
    .groupBy("store_id")
    .agg(
        sum("revenue").alias("Total Revenue")
    )
    .orderBy("store_id")
)
display(pipeline_df)
````

````python
# Displaying the pipeline schema
pipeline_df.printSchema()
````

````python
# Counting total stores
print("Total Stores:", pipeline_df.count())
````
