# Question 3: Handling Missing Values

```python
df.isnull().sum()
```

```python
# Missing discount values are assumed to indicate no discount.
df["discount"] = df["discount"].fillna(0)
```

```python
# Missing customer reviews indicate that no review information is available.
df["what_customers_said"] = df["what_customers_said"].fillna("No Reviews")
```

```python
# Missing seller names are replaced with 'Unknown' to retain all records.
df["seller_name"] = df["seller_name"].fillna("Unknown")
```

```python
# Missing video information indicates that no product video is available.
df["videos"] = df["videos"].fillna("No Video")
```

```python
# Missing seller details are replaced with 'Not Available'.
df["seller_information"] = df["seller_information"].fillna("Not Available")
```

```python
# Missing variation information indicates that no product variations are listed.
df["variations"] = df["variations"].fillna("No Variations")
```

```python
df.isnull().sum()
```

### Missing Value Handling Summary

Missing values were identified in six columns. Instead of removing records, missing values were replaced with meaningful defaults based on the context of each attribute. This approach preserves all 1000 records while ensuring that no null values remain in the dataset.

# Question 6: Create a derived column (total_amount = price * quantity)

### Since the dataset does not contain explicit `price` and `quantity` columns, I derived them from the available data:

- `price` was extracted from the `final_price` column by removing currency symbols and converting it to a numeric format.
- `quantity` was derived from the number of available size variants listed in the `sizes` column.
- Finally, `total_amount` was calculated as the product of price and quantity.

```python
df["price"] = (
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .astype(float)
)
```

```python
df[["final_price", "price"]].head()
```

```python
import ast
df["quantity"] = df["sizes"].apply(
    lambda x: len(ast.literal_eval(x))
)
```

```python
df[["sizes", "quantity"]].head()
```

```python
df["total_amount"] = df["price"] * df["quantity"]
```

```python
df[["price", "quantity", "total_amount"]].head()
```

### My Observation

A numeric `price` column was created from the `final_price` field. A `quantity` column was derived from the number of available size variants. Using these columns, a new feature named `total_amount` was created to represent the total value of a product based on its price and available quantity options.
